```python
import re

# code to make aspect and opinion word pairs
fullposlist = []

fpos = open('postags.txt','r')
postags = fpos.readline()
while postags:        
    poslist = []
    
    poslist.append(re.findall(r"[\w]+", postags))
        
    fullposlist.append(poslist)
    
    postags = fpos.readline()
fpos.close()


# get dependencies list

fulldeplist = []

fdep = open('dependencyparsing.txt','r')
deppars = fdep.readline()
while deppars:
    #code for parsing dep
    deppars = deppars.replace('[','')
    deppars = deppars.replace(']','')

    striptxt = deppars.split(')),')
    
    relationlist = []
    
    for st in striptxt:
        relationlist.append(re.findall(r"[\w]+", st))
        
    fulldeplist.append(relationlist)

    deppars = fdep.readline()
fdep.close()
```
